# **Pandas Series**
A pandas Series is like a column in a table. It is a 1-D array holding data of any type

# **Import Pandas**

```python
import pandas as pd
```

## **Series from lists**

```python
# string
country = ['Usa', 'Russia', 'UK', 'Europe']
pd.Series(country)
```

```python
# integers
data = [1,2,3,4,5,6]
pd.Series(data)
```

```python
# custome index
marks = [34,56,78,90]
student = ['A', 'B', 'C', 'D']
pd.Series(marks, index=student)
```

```python
# setting a name
pd.Series(marks, index=student, name='Marks')
```
