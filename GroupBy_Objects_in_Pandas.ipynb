```python
# last
genres.last()
```

```python
# nth
genres.nth(6)
```

```python
# get_group
genres.get_group('Action')
```

```python
# groups
genres.groups
```

```python
# describe
genres.describe()
```

```python
# sample
genres.sample()
```

```python
# nunique
genres.nunique()
```

```python
# agg method
genres.agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating': 'max',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'mean'
    }
)
```

```python
# passing list
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Metascore']].agg(['max','min','mean'])
```

```python
# adding both the syntax
genres.agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating': ['min','max'],
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'mean'
    }
)
```

```python
# looping on groups
df_list = []
for group, data in genres:
  df_list.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
df = pd.concat(df_list)
df
```

```python
# apply
genres.apply(min)
```

```python
# find the number of movies starting with A for each group
def foo(group):
  print(group)
  return group
```

```python
genres.apply(foo)
```

```python
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find the ranking of each movie in the group according to IMDB score
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find the normalized IMDB rating group wise
def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

# **GroupBy on Multiple Columns**

```python
# size
duo = movies.groupby(['Director','Star1'])
duo.size()
```

```python
# get_group
duo.get_group(('Christopher Nolan','Christian Bale'))
```

```python
# find the most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in term of metascore(avg)) actor -> genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
movies['Released_Year'] = pd.to_numeric(movies['Released_Year'], errors='coerce')
duo = movies.groupby(['Director','Star1'])
duo[['Released_Year', 'Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Metascore']].agg(['min','max','mean'])
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.head()
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# find the batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

```python
# find the V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# create a function that can return the highest score of any batsman
temp_df = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
temp_df
```
