# **Filtering DataFrame**

```python
ipl.head(5)
```

```python
# find all the final winners
final = ipl['MatchNumber'] == 'Final'
final_match = ipl[final]
final_match[['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
super = ipl['SuperOver'] == 'Y'
super_over = ipl[super]
super_over
```

```python
# how many matches has csk won in kolkata
csk = (ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')
csk_matches = ipl[csk]
csk_matches.shape[0]
```

```python
# toss winner is match winner in percentage
to_offset = ipl['TossWinner'] == ipl['WinningTeam']
to_offset_matches = ipl[to_offset]
(to_offset_matches.shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes > 10000
movies.head()
```

```python
moves = ((movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000))
moves = movies[moves]
moves
```

```python
# action movies with rating higher than 7.5
rating = ((movies['imdb_rating'] > 7.5) & (movies['genres'].str.split('|').apply(lambda x: 'Action' in x)))
rating = movies[rating]
rating
```

```python
# write a function than can return the track record of 2 teams against each other
track_record = ipl[(ipl['Team1'] == 'Chennai Super Kings') & (ipl['Team2'] == 'Kings XI Punjab')]
track_record
```
