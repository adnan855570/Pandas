```python
import pandas as pd
import numpy as np
```

```python
courses = pd.read_csv('/content/courses.csv')
courses.head()
```

```python
deliveries = pd.read_csv('/content/deliveries.csv')
deliveries.head()
```

```python
matches = pd.read_csv('/content/matches.csv')
matches.head()
```

```python
reg1 = pd.read_csv('/content/reg-month1.csv')
reg1.head()
```

```python
reg2 = pd.read_csv('/content/reg-month2.csv')
reg2.head()
```

```python
students = pd.read_csv('/content/students.csv')
students.head()
```

```python
# pd.concat
regs = pd.concat([reg1,reg2], ignore_index=True)
regs
```

```python
# concat dataframes horizontally
multi = pd.concat([reg1,reg2], keys=['Nov','Dec'])
multi
```

```python
multi.loc['Dec']
```

```python
multi.loc['Dec',5]
```

```python
pd.concat([reg1,reg2], axis=1)
```

```python
# inner join
students.merge(regs, how='inner', on='student_id')
```

```python
# left join
courses.merge(regs, how='left', on='course_id')
```

```python
# right join
temp_df = pd.DataFrame({
    'student_id': [26,27,28],
    'name' : ['Nitish','Ankit','Rahul'],
    'partner' : [28,26,17]
})

students = pd.concat([students,temp_df], ignore_index=True)
students
```

```python
students.merge(regs, how='right', on='student_id')
```

```python
# outer join
students.merge(regs, how='outer', on='student_id')
```

```python
# final the total revenue generated
total_revenue = regs.merge(courses, how='inner', on='course_id')['price'].sum()
total_revenue
```

```python
# find month by month revenue
temp_df = pd.concat([reg1,reg2],keys=['Nov','Dec']).reset_index()
temp_df
temp_df.merge(courses, on='course_id').groupby('level_0')['price'].sum()
```

```python
# print the registration table
regs.merge(students, on='student_id').merge(courses, on='course_id')[['name','course_name','price']]
```

```python
# plot bar chart for revenue/course
regs.merge(courses, on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# find students who enrolled in both the months
common_student_id = np.intersect1d(reg1['student_id'],reg2['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# find course that got no enrollment
course_list = np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_list)]
```

```python
# find the students who did not enroll into any course
student_list = np.setdiff1d(students['student_id'],regs['student_id'])
students[students['student_id'].isin(student_list)]
```

```python
# print student name -> partner name for all enrolled students
# self join
students.merge(students, left_on='partner', right_on='student_id')[['name_x','name_y']]
```

```python
# find top 3 students who did most number of enrollments
regs.merge(students, on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# find top 3 students who spent most amount of money on courses
regs.merge(students, on='student_id').merge(courses, on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
temp_df = deliveries.merge(matches, left_on='match_id', right_on='id')
temp_df
```

```python
num_matches = matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
# find the orange cap holder of all the seasons
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates('season').sort_values('season')
```
